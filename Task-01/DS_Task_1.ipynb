```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dataset.csv', skiprows=4)

df.drop(columns=['Unnamed: 69'], inplace=True)

df.dropna(subset=['2022'], inplace=True)

# --- Create the Histogram ---
plt.figure(figsize=(10, 7)) # Set a good figure size

sns.histplot(data=df, x='2022', kde=True, bins=30)

plt.yscale('log')

#--- Add Titles and Labels ---
plt.title('Distribution of World Country Populations in 2022 (Log Scale)', fontsize=16)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Number of Countries (Log Scale)', fontsize=12)
plt.grid(axis='y', alpha=0.5)

#--- Show and Save the Plot ---
plt.show()

plt.savefig('population_distribution_2022.png')
```
